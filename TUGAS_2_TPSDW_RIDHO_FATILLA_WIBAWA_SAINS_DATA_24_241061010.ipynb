### 1. Menampilkan Jumlah Data Hilang

Lakukan pemeriksaan terhadap dataset untuk mengetahui jumlah data hilang (missing values) pada setiap variabel.

Instruksi:

•	Tampilkan jumlah missing value pada setiap kolom.

•	Jelaskan variabel mana yang memiliki data hilang paling banyak.


```python
# Import library
import pandas as pd
import numpy as np

# Membaca file dataset csv
sd = pd.read_csv('/content/data_tugas 2.csv')

# Menampilkan 7 data teratas
sd.head(7)
```

Pada langkah ini dilakukan proses awal dalam data preparation, yaitu mengimpor library, membaca dataset, dan menampilkan sebagian data.

- Pandas digunakan untuk mengolah data dalam bentuk tabel (DataFrame), seperti membaca file dan menampilkan data.
- NumPy digunakan untuk operasi numerik serta membantu dalam penanganan nilai kosong seperti NaN.

Dataset dibaca menggunakan fungsi read_csv() dan disimpan dalam variabel sd. Selanjutnya, ditampilkan 7 baris pertama menggunakan head(7) untuk melihat struktur data.

Dari hasil tampilan, dapat ditemukan adanya nilai "?" pada beberapa variabel. Nilai tersebut menunjukkan data yang hilang, namun belum dikenali sebagai missing value oleh pandas. Oleh karena itu, nilai "?" perlu dikonversi menjadi NaN agar dapat diproses pada tahap berikutnya.

```python
# Mengganti tanda '?' menjadi NaN (missing value)
sd.replace('?', np.nan, inplace=True)

# Mengecek jumlah missing value pada setiap kolom
missing_values = sd.isnull().sum()

# Menampilkan Tabel Jumlah Missing Values
print("Tabel Jumlah Missing Values Pada Setiap Kolom:")
print(missing_values)
```

**Variabel dengan jumlah missing value terbanyak adalah normalized-losses sebanyak 41 data.** Hal ini menunjukkan bahwa variabel tersebut memiliki tingkat ketidaklengkapan data yang paling tinggi dan berpotensi mempengaruhi hasil analisis jika tidak ditangani dengan baik.

Selanjutnya, variabel jumlah-pintu memiliki 18 data hilang, diikuti oleh:

### 2. Mengganti Missing Value pada Variabel Kategori

Untuk variabel berbentuk kategori, data hilang harus diganti dengan kategori baru yaitu "unknown".

Variabel kategorik pada dataset:

•	make

•	bahan-bakar

•	jumlah-pintu

•	body-style

Instruksi:

•	Identifikasi missing value pada variabel kategori

•	Ganti nilai tersebut dengan "unknown"


```python
# Daftar variabel kategorik
kategori = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

# Mengganti missing value dengan 'unknown'
for col in kategori:
    sd[col] = sd[col].fillna('unknown')

# Menampilkan ringkasan jumlah setiap kategori
for col in kategori:
    print(f"\nTabel kategori untuk kolom: {col}")
    print(sd[col].value_counts())
```

Tabel ringkasan menunjukkan jumlah kemunculan setiap kategori pada variabel kategorik setelah missing value diganti dengan "unknown".

- Pada variabel make, kategori "unknown" berjumlah 4 data, menunjukkan jumlah data yang sebelumnya hilang.
- Pada bahan-bakar, kategori "unknown" berjumlah 13 data, sedangkan kategori terbanyak adalah "gas" (173 data).
- Pada jumlah-pintu, kategori "unknown" berjumlah 18 data, dengan kategori utama "four" (101 data) dan "two" (86 data).
- Pada body-style, kategori "unknown" berjumlah 12 data, dengan kategori terbanyak "sedan" (88 data).

Nilai-nilai tersebut menunjukkan distribusi data pada setiap kategori serta jumlah data yang sebelumnya hilang.

### 3. Mengganti Missing Value pada Variabel Numerik

Untuk variabel numerik yang memiliki data hilang, lakukan imputasi menggunakan nilai rata-rata (mean).

Variabel numerik pada dataset:

•	length

•	width

•	height

Instruksi:

•	Hitung nilai rata-rata masing-masing variabel

•	Gunakan nilai tersebut untuk mengganti data yang hilang.


```python
# Daftar variabel numerik
numerik = ['length', 'width', 'height']

# Menghitung mean dan mengganti missing value
for col in numerik:
    mean_value = sd[col].astype(float).mean()
    sd[col] = sd[col].astype(float).fillna(mean_value)
    print(f"Mean {col}: {mean_value}")
```

```python
sd.head(10)
```

Output yang ditampilkan merupakan nilai rata-rata (mean) dari masing-masing variabel numerik, yaitu:

- length ≈ 173.90
- width ≈ 65.93
- height ≈ 53.74

Nilai mean tersebut digunakan untuk menggantikan data yang hilang pada masing-masing variabel.

Dengan menggunakan nilai rata-rata, data yang kosong dapat diisi dengan nilai yang mewakili kondisi umum dari data tersebut, sehingga tidak mengganggu distribusi data secara signifikan.

### 4. Min-Max Normalization
Lakukan Min-Max Normalization pada variabel:

wheel-bases


```python
# Menghitung nilai minimum dan maksimum
min_val = sd['wheel-bases'].astype(float).min()
max_val = sd['wheel-bases'].astype(float).max()

# Melakukan Min-Max Normalization
sd['wheel-bases-normalized'] = (sd['wheel-bases'].astype(float) - min_val) / (max_val - min_val)

# Menampilkan hasil
sd[['wheel-bases', 'wheel-bases-normalized']].head()
```

Pada langkah ini dilakukan normalisasi data menggunakan metode Min-Max Normalization pada variabel wheel-bases.

Metode ini mengubah nilai data ke dalam rentang 0 hingga 1:
- Rumus Min-Max Normalization

$$
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

- min() digunakan untuk mencari nilai minimum
- max() digunakan untuk mencari nilai maksimum
- Hasil normalisasi disimpan dalam kolom baru yaitu wheel-bases-normalized

Nilai pada kolom wheel-bases-normalized merupakan hasil transformasi dari nilai asli menggunakan metode Min-Max Normalization.

- Nilai 88.6 diubah menjadi 0.058309

- Nilai 94.5 diubah menjadi 0.230321

- Nilai 99.8 diubah menjadi 0.384840

Perubahan ini terjadi karena setiap nilai disesuaikan berdasarkan nilai minimum dan maksimum dari seluruh data.
Hasil output menunjukkan bahwa variabel wheel-bases telah berhasil dinormalisasi ke dalam skala 0–1 melalui kolom baru wheel-bases-normalized, sehingga memudahkan proses analisis dan pemodelan selanjutnya.

### 5. Z-Score Normalization

Lakukan Z-score normalization pada variabel berikut:

•	length

•	width

•	height


```python
# Daftar variabel numerik
numerik = ['length', 'width', 'height']

# Melakukan Z-Score Normalization
for col in numerik:
    mean = sd[col].astype(float).mean()
    std = sd[col].astype(float).std()

    sd[col + '-zscore'] = (sd[col].astype(float) - mean) / std

# Menampilkan hasil
sd[['length', 'length-zscore',
    'width', 'width-zscore',
    'height', 'height-zscore']].head()
```

Pada langkah ini dilakukan normalisasi data menggunakan metode Z-Score pada variabel length, width, dan height.

Metode Z-Score digunakan untuk mengubah data sehingga memiliki:

- Rata-rata (mean) ≈ 0
- Standar deviasi (std) ≈ 1

### Rumus Z-Score Normalization

$$
Z = \frac{X - \mu}{\sigma}
$$


Hasil normalisasi disimpan dalam kolom baru:

- length-zscore
- width-zscore
- height-zscore

Output menampilkan nilai asli dan hasil normalisasi Z-score.

- Nilai positif → di atas rata-rata
- Nilai negatif → di bawah rata-rata
- Nilai mendekati 0 → dekat dengan rata-rata

Dengan normalisasi ini, data menjadi lebih terstandarisasi tanpa mengubah pola distribusinya.

Variabel numerik telah berhasil dinormalisasi menggunakan metode Z-Score, sehingga data memiliki skala yang seragam dan lebih optimal untuk analisis maupun pemodelan.
