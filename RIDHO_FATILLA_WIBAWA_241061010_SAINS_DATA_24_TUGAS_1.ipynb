4. jelaskan struktur dataset tersebut (jumlah baris, kolom, dan tipe data)

Dataset customers.csv memiliki struktur data yang terdiri dari 2000 baris (entries) dan 8 kolom (variabel). Setiap baris merepresentasikan satu data pelanggan, sedangkan setiap kolom berisi atribut informasi pelanggan seperti CustomerID, Gender, Age, Annual Income ($), Spending Score (1–100), Profession, Work Experience, dan Family Size. Berdasarkan tipe datanya, dataset ini memiliki dua jenis tipe data yaitu int64 untuk data numerik (CustomerID, Age, Annual Income, Spending Score, Work Experience, dan Family Size) serta object untuk data kategorik atau teks (Gender dan Profession). Struktur ini menunjukkan bahwa dataset berisi kombinasi data numerik dan kategorikal yang dapat digunakan untuk analisis data pelanggan.

5. buatkan struktur data dengan format json, secara manual dan konversi menjadi dataset dalam format csv (dataset hasil konversi ke csv di upload di github)

```python
# 1. Data JSON
data = [
    {"ID": 1, "Nama": "Andi", "Umur": 30, "Kota": "Makassar"},
    {"ID": 2, "Nama": "Siti", "Umur": 32, "Kota": "Jakarta"},
    {"ID": 3, "Nama": "Budi", "Umur": 31, "Kota": "pare-pare"}
]

# 2. Simpan sebagai file JSON
with open("data.json", "w") as file:
    json.dump(data, file, indent=4)

# 3. Konversi JSON ke DataFrame
df = pd.DataFrame(data)

# 4. Simpan menjadi CSV
df.to_csv("data_peserta.csv", index=False)

print("Konversi JSON ke CSV berhasil!")
```
