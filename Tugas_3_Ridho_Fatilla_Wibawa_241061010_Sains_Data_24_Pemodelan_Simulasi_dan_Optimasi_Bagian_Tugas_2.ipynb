# Sebuah restoran cepat saji memiliki rata-rata 5 pelanggan datang setiap 10 menit.



```python
import numpy as np
import matplotlib.pyplot as plt

customers_per_10min = 5

lam = customers_per_10min * 6  # 60/10 = 6

print(lam)
```

### Perhitungan Parameter λ (Lambda)

Diketahui:
- Rata-rata pelanggan = 5 pelanggan per 10 menit  
- 1 jam = 60 menit = 6 interval  

Maka:

$$
\lambda = 5 \times 6 = 30
$$

Sehingga rata-rata jumlah pelanggan per jam adalah **30 pelanggan**.

## 1. Gunakan distribusi Poisson untuk mensimulasikan jumlah pelanggan yang datang dalam satu jam (60 menit).

```python
# Parameter
customers_per_10min = 5

# Hitung lambda per jam
lam = customers_per_10min * 6  # 60/10 = 6

# Simulasi 1 jam (1 data saja)
customers_1hour = np.random.poisson(lam)

print("Lambda (rata-rata per jam):", lam)
print("Jumlah pelanggan dalam 1 jam (hasil simulasi):", customers_1hour)
```

 Berdasarkan simulasi menggunakan distribusi Poisson, diperoleh jumlah pelanggan yang datang dalam satu jam adalah **24 pelanggan**. Hasil simulasi sebesar **24** masih wajar karena berada di sekitar nilai rata-rata (**λ = 30**), mengingat distribusi Poisson bersifat acak.

## 2. Jalankan simulasi untuk 100 hari dan visualisasikan hasilnya dalam histogram.

```python
# Supaya hasil konsisten
np.random.seed(42)

# Parameter
lam = 30   # rata-rata pelanggan per jam
days = 100  # jumlah hari simulasi

# Simulasi 100 hari
customers = np.random.poisson(lam, days)

print("Data simulasi:\n", customers)

```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(customers, bins=15, edgecolor='black')

plt.xlabel("Jumlah Pelanggan per Jam")
plt.ylabel("Frekuensi")
plt.title("Histogram Jumlah Pelanggan Selama 100 Hari (Poisson)")

plt.grid(True)
plt.show()
```

Simulasi jumlah pelanggan dilakukan selama 100 hari menggunakan distribusi Poisson dengan parameter λ = 30 pelanggan per jam.

Berdasarkan hasil simulasi, diperoleh data jumlah pelanggan yang bervariasi, dengan nilai berkisar antara sekitar **17** hingga **43** pelanggan per jam.

Histogram menunjukkan bahwa sebagian besar jumlah pelanggan berada di kisaran **25** hingga **35** pelanggan per jam, dengan puncak distribusi di sekitar nilai rata-rata yaitu **30**.

```python
#Bar chart
plt.figure(figsize=(10, 5))
plt.bar(range(1, days+1), customers, color="b", alpha=0.7)

plt.xlabel("Hari")
plt.ylabel("Jumlah Pelanggan per Jam")
plt.title("Simulasi Jumlah Pelanggan Restoran (Distribusi Poisson)")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

Selain menggunakan histogram, hasil simulasi juga divisualisasikan dalam bentuk diagram batang (bar chart) untuk menunjukkan jumlah pelanggan per jam selama **100** hari.

Grafik menunjukkan fluktuasi jumlah pelanggan setiap hari, dengan nilai yang bervariasi di sekitar rata-rata **30** pelanggan per jam. Grafik batang menunjukkan bahwa jumlah pelanggan berubah-ubah setiap hari, namun tetap berada di sekitar nilai rata-rata. Hal ini mengindikasikan bahwa kedatangan pelanggan bersifat acak, tetapi masih mengikuti pola distribusi Poisson.

Dengan adanya histogram dan grafik batang, dapat disimpulkan bahwa hasil simulasi tidak hanya menunjukkan distribusi frekuensi, tetapi juga pola fluktuasi harian yang realistis.


## 3. Hitung rata-rata dan variansi jumlah pelanggan per jam berdasarkan hasil simulasi.

```python
mean_customers = np.mean(customers)
variance_customers = np.var(customers)

print("Rata-rata:", mean_customers)
print("Variansi:", variance_customers)
```

Berdasarkan hasil simulasi selama **100 hari**, diperoleh rata-rata jumlah pelanggan per jam sebesar **30.05** pelanggan, sedangkan variansi sebesar **28.95**.

Nilai rata-rata dan variansi yang diperoleh mendekati nilai parameter **λ = 30**. Hal ini sesuai dengan sifat distribusi Poisson, di mana nilai mean dan variansi secara teoritis sama dengan λ. Perbedaan kecil antara hasil simulasi dengan nilai teoritis terjadi karena adanya unsur acak dalam simulasi. Namun, hasil yang diperoleh tetap menunjukkan bahwa distribusi Poisson merupakan model yang sesuai untuk menggambarkan jumlah kedatangan pelanggan per jam.

Semakin banyak jumlah simulasi yang dilakukan, maka hasil rata-rata dan variansi akan semakin mendekati nilai teoritis (λ).

## 4. Jika restoran memiliki kapasitas maksimal 40 pelanggan per jam, berapa probabilitas terjadinya kelebihan kapasitas?

```python
prob_over_capacity = np.sum(customers > 40) / len(customers)

print("Probabilitas > 40 pelanggan:", prob_over_capacity)
```

Berdasarkan hasil simulasi, probabilitas jumlah pelanggan yang melebihi kapasitas **40** pelanggan per jam adalah sebesar **0.03** atau **3%**.

Hal ini menunjukkan bahwa kejadian kelebihan kapasitas terjadi dengan peluang yang relatif kecil. Dengan probabilitas hanya **3%**, dapat disimpulkan bahwa kapasitas restoran sebesar **40** pelanggan per jam masih mencukupi untuk melayani pelanggan dalam sebagian besar kondisi. Risiko terjadinya kelebihan kapasitas tergolong rendah.
