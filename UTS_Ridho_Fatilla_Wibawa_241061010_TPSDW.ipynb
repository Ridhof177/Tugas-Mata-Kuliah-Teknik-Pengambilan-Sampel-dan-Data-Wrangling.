### 1.	Lakukan scraping web (https://deepublishstore.com/product-category/ekonomi/page/1/?) di google colab, scraping halaman 1- 4, dengan mengambil informasi
-	Judul Buku (code: ("h2", class_="woocommerce-loop-product__title"))
-	Harga Buku (code: ("span", class_="woocommerce-Price-amount amount"))
-	Link Gambar Buku (code: ("img", class_="attachment-woocommerce_thumbnail size-woocommerce_thumbnail"))


```python
# Langkah 1
import requests
from bs4 import BeautifulSoup
import json
import time

url = "https://deepublishstore.com/product-category/ekonomi/page/1/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Menampilkan struktur HTML
print(soup.prettify())
```

**Langkah pertama** dalam proses web scraping adalah menyiapkan lingkungan kerja yang diperlukan untuk mengambil dan mengolah data dari website. Tahap ini meliputi:

1. Mengimpor library yang dibutuhkan.
2. Menentukan URL target yang akan di-scrape.
3. Menambahkan header (User-Agent) untuk mensimulasikan permintaan dari browser.
4. Mengirim permintaan HTTP ke server website.
5. Melakukan parsing HTML menggunakan BeautifulSoup agar struktur halaman dapat dianalisis.

```python
# Langkah 2
# Mengecek ketersediaan informasi pada HTML

print("woocommerce-loop-product__title" in soup.prettify())
print("woocommerce-Price-amount amount" in soup.prettify())
print("attachment-woocommerce_thumbnail size-woocommerce_thumbnail" in soup.prettify())
```

**Langkah kedua** digunakan untuk memastikan bahwa class HTML yang menjadi target scraping действительно tersedia pada halaman web.
Jika hasilnya True, maka elemen tersebut dapat diambil.
Jika False, maka kemungkinan class salah atau struktur HTML telah berubah.

```python
# Langkah 3: Scraping data dari halaman 1 sampai 4
List_Buku = []

base_url = "https://deepublishstore.com/product-category/ekonomi/page/{}/"

for page in range(1, 5):  # Halaman 1 hingga 4
    print(f"Scraping halaman {page}...")

    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Mengambil data berdasarkan class
    judul = soup.find_all("h2", class_="woocommerce-loop-product__title")
    harga = soup.find_all("span", class_="woocommerce-Price-amount amount")
    gambar = soup.find_all(
        "img",
        class_="attachment-woocommerce_thumbnail size-woocommerce_thumbnail"
    )

    # Menyamakan jumlah data agar tidak terjadi error
    min_len = min(len(judul), len(harga), len(gambar))

    for i in range(min_len):
        data_buku = {
            "judul_buku": judul[i].get_text(strip=True),
            "harga_buku": harga[i].get_text(strip=True),
            "link_gambar": gambar[i]["src"] if gambar[i].has_attr("src") else "N/A"
        }
        List_Buku.append(data_buku)

    # Delay untuk menghindari beban berlebih pada server
    time.sleep(2)
```

**Langkah 3** merupakan tahap inti dalam proses web scraping, yaitu mengambil (mengekstraksi) informasi yang dibutuhkan dari halaman web. Pada tugas ini, proses ekstraksi dilakukan pada website Deepublish Store untuk kategori buku Ekonomi dari halaman 1 hingga 4.

Informasi yang diambil meliputi:

- Judul Buku
- Harga Buku
- Link Gambar Buku

Data tersebut kemudian disimpan dalam sebuah struktur data berbentuk list of dictionaries yang diberi nama List_Buku.

```python
# Langkah 4: Menampilkan beberapa hasil scraping
for buku in List_Buku:
    print(buku)

print("Total data buku:", len(List_Buku))
```

Langkah 4 – Menampilkan dan Verifikasi Hasil Scraping

Pada tahap ini dilakukan verifikasi terhadap data hasil scraping yang telah disimpan dalam variabel List_Buku. Proses verifikasi dilakukan dengan menampilkan setiap elemen data menggunakan perulangan serta menghitung jumlah total data yang berhasil dikumpulkan menggunakan fungsi len(). Tahap ini bertujuan untuk memastikan bahwa proses ekstraksi data dari halaman 1 hingga 4 telah berjalan dengan baik dan menghasilkan data yang terstruktur serta siap untuk disimpan dalam format JSON.

### 2.	Berikan nama hasil scraping bernama List_Buku, ubah format file menjadi JSON

```python
# Langkah 5: Konversi ke format JSON
json_data = json.dumps(List_Buku, indent=4, ensure_ascii=False)

# Menyimpan ke file JSON
with open("List_Buku.json", "w", encoding="utf-8") as file:
    file.write(json_data)

print("File 'List_Buku.json' berhasil disimpan.")
```

Langkah 5 – Konversi dan Penyimpanan Data ke Format JSON

Pada tahap ini, data hasil scraping yang tersimpan dalam variabel List_Buku dikonversi ke dalam format JSON menggunakan fungsi json.dumps(). Format JSON dipilih karena bersifat ringan, mudah dibaca, dan kompatibel dengan berbagai sistem. Selanjutnya, data tersebut disimpan dalam file bernama List_Buku.json dengan encoding UTF-8 untuk memastikan karakter ditampilkan dengan benar. Tahap ini menandai selesainya proses web scraping dan memastikan bahwa data dapat digunakan kembali untuk analisis atau integrasi lebih lanjut.

### 4.	Ubah format data menjadi csv menggunakan prinsip ETL. Upload hasil ETL dalam bentuk csv ke github

### - Extract (Ekstraksi Data)

```python
# Extract: Membaca data dari file JSON
with open("List_Buku.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

Tahap Extract merupakan proses pengambilan data dari sumber data yang tersedia. Dalam penelitian ini, data diperoleh melalui proses web scraping dari website Deepublish Store pada kategori buku Ekonomi halaman 1 hingga 4. Data yang berhasil dikumpulkan meliputi judul buku, harga buku, dan link gambar buku. Hasil ekstraksi disimpan dalam sebuah struktur data berbentuk list of dictionaries dengan nama List_Buku, serta dapat disimpan dalam format JSON untuk memudahkan proses selanjutnya.
di mana proses ekstraksi bertujuan untuk mengumpulkan data mentah dari berbagai sumber sebelum dilakukan pengolahan lebih lanjut.

```python
import pandas as pd

# Transform: Membuat DataFrame
df = pd.DataFrame(data)

# Membersihkan kolom harga
df['harga_buku'] = (
    df['harga_buku']
    .str.replace('Rp', '', regex=False)
    .str.replace('.', '', regex=False)
    .astype(int)
)

# Menghapus data duplikat
df.drop_duplicates(inplace=True)

# (Opsional) Mengganti nama kolom agar lebih ringkas
df.rename(columns={
    'judul_buku': 'judul',
    'harga_buku': 'harga',
    'link_gambar': 'gambar'
}, inplace=True)

# Menampilkan hasil transformasi
print(df.head())
```

### - Transform (Transformasi Data)

**Tahap Transform** bertujuan untuk membersihkan dan mengolah data agar siap digunakan untuk analisis atau penyimpanan. Pada tahap ini dilakukan beberapa proses, antara lain:

1. **Konversi data ke dalam bentuk DataFrame** menggunakan library pandas.
2. **Pembersihan format harga,** yaitu menghapus simbol “Rp” dan tanda titik (.) sehingga dapat diubah menjadi tipe data numerik.
3. **Penghapusan data duplikat** untuk menjaga kualitas data.
4. **Penyesuaian nama kolom** agar lebih ringkas dan mudah dipahami (opsional).


### - Load (Pemuatan Data)

```python
# Load: Menyimpan ke file CSV
df.to_csv("List_Buku.csv", index=False, encoding="utf-8-sig")

print("File 'List_Buku.csv' berhasil disimpan.")
```

**Tahap Load** merupakan proses penyimpanan data yang telah ditransformasi ke dalam media penyimpanan tertentu. Dalam penelitian ini, data disimpan dalam format CSV dengan nama List_Buku.csv.
